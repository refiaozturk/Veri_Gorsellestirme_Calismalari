# Kütüphaneleri Yükleme

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings;
warnings.filterwarnings("ignore")
```

# Değişkenleri Tanımlama

Grafik oluştururken kullanacağımız m, c ve E değişkenlerini tanımlayalım.

```python
# m değişkenini numpy kullanarak tanımlama
m = np.arange(0, 11, dtype= "float")
m
```

```python
# c değişkenini tanımlama
c = 3 * 10 ** 8
```

```python
# m ve c değişekenlerini bir matematik formülünde kullanarak E değişkenini tanımlama
E = m * c ** 2
E
```

# Grafik Oluşturma

## E = mc^2 Grafiğini Çizdirme

```python
plt.figure(figsize = (6,4)) # Figsize argümanı figürün boyutunu ayarlamak için kullanılır.

# Grafiği çizdirme
plt.plot(m, E, "b", lw= 7) # "b" blue yani mavi anlamına gelmektedir, çizginin rengini mavi yapar. lw, linewidth yani çizgi kalınlığı demektir.

# x, y eksenlerini ve grafik başlığını yazma
plt.xlabel("m")
plt.ylabel("E")
plt.title("E = mc^2")

plt.show()
```

```python
# Farklı renk ve farklı çizgi kalınlığı ile E= mc^2 grafiğini oluşturma

plt.figure(figsize = (6,4))

plt.plot(m, E, "orange", lw= 3)

plt.xlabel("m")
plt.ylabel("E")
plt.title("E = mc^2")

plt.show()
```

```python
# Farklı renk ve farklı figür boyutunda logaritmik E= mc^2 grafiği oluşturma

plt.figure(figsize = (8, 5))
plt.plot(m, E, "g", lw= 3) # "g" green demektir ve çizgi rengini yeşil yapar.

plt.xlabel("m", size = 15) # size argümanı x ekseni etiketinin boyutunu ayarlamaya yardımcı olur. 
plt.ylabel("E", size = 15) # size argümanı y ekseni etiketinin boyutunu ayarlamaya yardımcı olur. 
plt.title("E = mc^2")

plt.yscale("log") # çizgiyi logaritmik hale getirir.

plt.grid(which = 'both', linestyle = '-', linewidth = 0.5, axis = "y") # grafiğin arkasına grid ekler.

plt.show()
```

# Yeni Değişkenler Tanımlama

```python
# yeni değişkenler

labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

# Yeni Değişkenlerle Grafik Oluşturma

```python
plt.figure(figsize = (6, 4))

plt.plot(labels, july16_2007, "steelblue", label = "july16_2007")
plt.plot(labels, july16_2020, "orange", label = "july16_2020")

plt.legend(bbox_to_anchor = (1, 0.55))

plt.show()
```

```python
plt.figure(figsize = (6, 4))

plt.plot(labels, july16_2007, "steelblue", label = "july16_2007")
plt.plot(labels, july16_2020, "orange", label = "july16_2020")

plt.legend(bbox_to_anchor = (1.33, 0.55))

plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize = (7, 6))

ax[0].plot(labels, july16_2007, "steelblue", lw = 2)
ax[0].set_title("july16_2007")

ax[1].plot(labels, july16_2020, "steelblue", lw = 2)
ax[1].set_title("july16_2020")

plt.tight_layout()

plt.show()
```

```python
# y eksenlerinde farklı değerler kullanma

fig, ax1 = plt.subplots(figsize = (8.5, 6))

ax1.plot(labels, july16_2007, 'b-', label='July 16 2007')
ax1.set_ylabel('July 16 2007', color='blue', fontsize= 15)
ax1.tick_params(axis='y', labelsize= 8)

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, 'r-', label='July 16 2020')
ax2.set_ylabel('July 16 2020', color='red', fontsize= 15)
ax2.tick_params(axis='y', labelsize= 8)

plt.show()
```

```python
# ax1 ve ax2 için y değerlerinin renklerini değiştirme

fig, ax1 = plt.subplots(figsize = (8.5, 6))

ax1.plot(labels, july16_2007, 'b-', label='July 16 2007')
ax1.set_ylabel('July 16 2007', color='blue', fontsize= 15)
ax1.tick_params(axis='y', labelsize= 8)
plt.yticks(size=7, color="blue")

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, 'r-', label='July 16 2020')
ax2.set_ylabel('July 16 2020', color='red', fontsize= 15)
ax2.tick_params(axis='y', labelsize= 8)
plt.yticks(size=7, color="red")

plt.show()
```
